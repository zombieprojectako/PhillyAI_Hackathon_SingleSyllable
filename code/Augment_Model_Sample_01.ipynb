```python
import pandas as pd
import numpy as np
import os
from datetime import datetime, date, time
```

# Load Data


```python
fp = os.path.join('..', 'data_out', 'model_sample_01.csv')
df = pd.read_csv(fp, parse_dates=['cycle_start_date'])
```

```python
df.info()
```

```python
def parse_mmdd(id, date_in): 
    tt = date_in.timetuple()
    month_str = str(tt[1])
    year_str = str(tt[0])
    rtn_str = str(id) + year_str + month_str
    return(rtn_str)
```

```python
df['key'] = ''
```

```python
df.head()
```

```python
for i in range(0, len(df)):
    df.iloc[i,17] = parse_mmdd(df.iloc[i, 0], df.iloc[i, 1])
    
```

```python
df.head()
```

```python
# Expected Cycle 
df.expected_cycle_length = pd.to_timedelta(df.expected_cycle_length)
df['est_cycle'] = df.expected_cycle_length.dt.days

# Actual Cycle 
df.cycle_length = pd.to_timedelta(df.cycle_length)
df['act_cycle'] = df.cycle_length.dt.days
```

```python
df.head()
```

```python
df = df.fillna(0)
```

```python
# Expected Period 
df.expected_period_length = pd.to_timedelta(df.expected_period_length)
df['est_period'] = df.expected_period_length.dt.days

# Actual Period 
df.mean_period_length = pd.to_timedelta(df.mean_period_length)
df['act_period'] = df.mean_period_length.dt.days
```

```python
df.head()
```

```python
fp = os.path.join('..', 'data_out', 'model_sample_02.csv')
df.to_csv(fp)
```
